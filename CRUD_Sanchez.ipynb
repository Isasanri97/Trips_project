# Creacion y conexion a BD

```python
import sqlite3
conn = sqlite3.connect('trips.db')
```

# Importe de datos

```python
import pandas as pd
data = '/content/drive/My Drive/trips_data.csv'
df = pd.read_csv(data)
trips = 'trips'
df.to_sql(trips, conn, if_exists='replace', index=False)
```

# CRUD

```python
#CREATE
cursor = conn.cursor()
new_trip=("AEIBK2024282","TS15UC9345","Kollur, Medak, Telangana","Lakdaram, Medak, Telangana","2020-08-01 17:44:00.000","2020-08-01 15:44:00.000", "delay","V0045778", "LEH04260", "2020-07-30 13:46:22", 86, "October","10")
cursor.execute(
    "INSERT INTO trips (BookingID,vehicle_no,Origin_Location,Destination_Location,Planned_ETA,actual_eta,ontime,OriginLocation_Code,DestinationLocation_Code,trip_start_date,TRANSPORTATION_DISTANCE_IN_KM,trip_month,trip_month_number) VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?)",new_trip)
conn.commit()
cursor.execute("SELECT * FROM trips WHERE BookingID = 'AEIBK2024282'")
new_data = cursor.fetchone()
print(new_data)
#conn.close()
```

```python
#READ
cursor = conn.cursor()
cursor.execute("SELECT * FROM trips")
registros = cursor.fetchall()

# Imprimir los registros
for registro in registros:
    print(registro)
#conn.close()
```

```python
#UPDATE
cursor = conn.cursor()
new_BookingID = "AEIBK20274155"
new_vehicle_no = "KA51D1317"
cursor.execute("UPDATE trips SET BookingID=? WHERE vehicle_no =?",(new_BookingID,new_vehicle_no))
conn.commit()
cursor.execute("SELECT * FROM trips WHERE BookingID = 'AEIBK20274155'")
update_register= cursor.fetchone()
print(update_register)
#conn.close()
```

```python
#DELETE
BookingID="AEIBK20274155"
cursor.execute("DELETE FROM trips WHERE BookingID=?",(BookingID,))
conn.commit()
cursor.execute("SELECT * FROM trips WHERE BookingID = 'AEIBK20274155'")
delete_register= cursor.fetchone()
if delete_register:
  print(delete_register)
else:
  print("not found")
#conn.close()
```
